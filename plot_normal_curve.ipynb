```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
samples_1 = np.random.normal(20, 1, size=100000)
samples_2 = np.random.normal(20, 3, size=100000)
samples_3 = np.random.normal(20, 10, size=100000)
```

```python
_ = plt.hist(samples_1, bins=100, density=True, histtype='step')
_ = plt.hist(samples_2, bins=100, density=True, histtype='step')
_ = plt.hist(samples_3, bins=100, density=True, histtype='step')

```
